```python
# Define the function to calculate temperature range for each farmer's field
def calculate_temperature_range(field_id):
    # Filter the data for the specific field_id
    field_data = data[data['Field_ID'] == field_id]
    # Calculate the temperature range
    temperature_range = field_data['Max_temperature_C'].iloc[0] - field_data['Min_temperature_C'].iloc[0]
    return temperature_range

# Call the function with the specified Field_IDs
temp_range_1458 = calculate_temperature_range(1458)
temp_range_1895 = calculate_temperature_range(1895)
temp_range_5443 = calculate_temperature_range(5443)
```

```python
# Extract all unique crop types from the dataset
unique_crop_types = data['Crop_type'].unique()

# Define a recursive function to sum the lengths of the crop names
def sum_crop_lengths(crop_list, total=0):
    if len(crop_list) == 0:  # Base case: no more crops in the list
        return total
    else:
        # Add the length of the first crop in the list to the total, then recurse with the rest of the list
        return sum_crop_lengths(crop_list[1:], total + len(crop_list[0]))

# Calculate the sum of the lengths of unique crop types
total_length = sum_crop_lengths(unique_crop_types)
total_length
```

```python
import pandas as pd
from scipy import stats

# Load the data
data = pd.read_csv('MD_agric_exam-4313.csv')  # Ensure the file path is correct

# Filter data for 'coffee' and 'banana' crop types
coffee_yield = data[data['Crop_type'] == 'coffee']['Annual_yield']
banana_yield = data[data['Crop_type'] == 'banana']['Annual_yield']

# Perform a t-test comparing the average 'Annual_yield'
t_stat, p_value = stats.ttest_ind(coffee_yield, banana_yield)

# Round the p-value to three decimal places
p_value_rounded = round(p_value, 3)

# Print the rounded p-value
print(p_value_rounded)
```
